<h2 align="middle">Milestone Status Report</h2>
<p><b>Prompt:</b> Summarize what you have accomplished</p>

<p>We have a basic simulation of the water particles up and running. We set up a basic scene where we applied gravitational force on the particles to replicate free fall. We are currently working on scene interaction along with self-collisions.</p>

<p><b>Prompt:</b> Preliminary results</p>

<p>We can now simulate gravitational effects on the particles and have the framework in place to integrate over applied forces in each tick of the simulation. In addition, we added a transparency effect to the particles in hopes that it would be easier to percieve depth and see what was going on with particles in the background. Lastly, we have a preliminary implementation of plane geometry (a plane is currently defined as four corners and a normal vector, we want to switch this to a mesh-like model for more complex scenes) </p>
<p>We have noticed that a large portion of our setbacks are trying to parse the coordinate system in our scene, which follows the convention in the lectures but is still kind of jarring to work with without much experience. Additionally, getting used to how we handle geometry in OpenGL and how to display different geometric designs in the render space has been tricky but we are getting the hang of it</p>

<p><b>Prompt:</b> Reflect on progress relative to your plan</p>

<p>We had some difficulties working out the finer details of compiling and running the simulation pipeline that was provided to us. After messing with some homebrew and making sure we had the right versions of all the libraries that we needed, we got it working and we have been building on it since.  We are currently working on scene intersection and self-collisions. These are key to having a realistic fluid simulation since a lot of a fluid's motion is defined by aggregate interaction between its constituent pars. We are currently a little behind schedule and are making progress towards finishing all the way through our Week 3 objectives although we should still be on track and won't need to modify our objectives.</p>
